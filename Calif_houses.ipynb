```python
#  Задание 44 
#  В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего из 1 столбца. 
#  Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без get_dummies?
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.loc[data['whoAmI'] == 'robot', 'for_machine_learning'] = '0'
data.loc[data['whoAmI'] == 'human', 'for_machine_learning'] = '1'

data.head(n=20)
```

```python
#  Задание 44 
#  В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего из 1 столбца. 
#  Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без get_dummies?
import random
lst = ['robot'] * 10
lst += ['human'] * 10
lst_1 = []  # Создаем новый список для конвертации строк из первого массива в вид "one hot"
random.shuffle(lst)
print(lst) # напечатаем изначальный рандомно сгенерированный список
for i in lst:
  if i == 'robot':
    i = 0
    lst_1.append(i)
  elif i == 'human':
    i = 1
    lst_1.append(i)
print(lst_1) # напечатаем "one hot" список

data = pd.DataFrame({'whoAmI':lst_1}) # выведем получившуюся таблицу
data.head(n=20)
```
