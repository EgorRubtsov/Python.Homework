```python
#  Ð—Ð°Ð´Ð°Ð½Ð¸Ðµ 44 
#  Ð’ ÑÑ‡ÐµÐ¹ÐºÐµ Ð½Ð¸Ð¶Ðµ Ð¿Ñ€ÐµÐ´ÑÑ‚Ð°Ð²Ð»ÐµÐ½ ÐºÐ¾Ð´ Ð³ÐµÐ½ÐµÑ€Ð¸Ñ€ÑƒÑŽÑ‰Ð¸Ð¹ DataFrame, ÐºÐ¾Ñ‚Ð¾Ñ€Ð°Ñ ÑÐ¾ÑÑ‚Ð¾Ð¸Ñ‚ Ð²ÑÐµÐ³Ð¾ Ð¸Ð· 1 ÑÑ‚Ð¾Ð»Ð±Ñ†Ð°. 
#  Ð’Ð°ÑˆÐ° Ð·Ð°Ð´Ð°Ñ‡Ð° Ð¿ÐµÑ€ÐµÐ²ÐµÑÑ‚Ð¸ ÐµÐ³Ð¾ Ð² one hot Ð²Ð¸Ð´. Ð¡Ð¼Ð¾Ð¶ÐµÑ‚Ðµ Ð»Ð¸ Ð²Ñ‹ ÑÑ‚Ð¾ ÑÐ´ÐµÐ»Ð°Ñ‚ÑŒ Ð±ÐµÐ· get_dummies?
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.loc[data['whoAmI'] == 'robot', 'for_machine_learning'] = '0'
data.loc[data['whoAmI'] == 'human', 'for_machine_learning'] = '1'

data.head(n=20)
```

```python
#  Ð—Ð°Ð´Ð°Ð½Ð¸Ðµ 44 
#  Ð’ ÑÑ‡ÐµÐ¹ÐºÐµ Ð½Ð¸Ð¶Ðµ Ð¿Ñ€ÐµÐ´ÑÑ‚Ð°Ð²Ð»ÐµÐ½ ÐºÐ¾Ð´ Ð³ÐµÐ½ÐµÑ€Ð¸Ñ€ÑƒÑŽÑ‰Ð¸Ð¹ DataFrame, ÐºÐ¾Ñ‚Ð¾Ñ€Ð°Ñ ÑÐ¾ÑÑ‚Ð¾Ð¸Ñ‚ Ð²ÑÐµÐ³Ð¾ Ð¸Ð· 1 ÑÑ‚Ð¾Ð»Ð±Ñ†Ð°. 
#  Ð’Ð°ÑˆÐ° Ð·Ð°Ð´Ð°Ñ‡Ð° Ð¿ÐµÑ€ÐµÐ²ÐµÑÑ‚Ð¸ ÐµÐ³Ð¾ Ð² one hot Ð²Ð¸Ð´. Ð¡Ð¼Ð¾Ð¶ÐµÑ‚Ðµ Ð»Ð¸ Ð²Ñ‹ ÑÑ‚Ð¾ ÑÐ´ÐµÐ»Ð°Ñ‚ÑŒ Ð±ÐµÐ· get_dummies?
import random
lst = ['robot'] * 10
lst += ['human'] * 10
lst_1 = []  # Ð¡Ð¾Ð·Ð´Ð°ÐµÐ¼ Ð½Ð¾Ð²Ñ‹Ð¹ ÑÐ¿Ð¸ÑÐ¾Ðº Ð´Ð»Ñ ÐºÐ¾Ð½Ð²ÐµÑ€Ñ‚Ð°Ñ†Ð¸Ð¸ ÑÑ‚Ñ€Ð¾Ðº Ð¸Ð· Ð¿ÐµÑ€Ð²Ð¾Ð³Ð¾ Ð¼Ð°ÑÑÐ¸Ð²Ð° Ð² Ð²Ð¸Ð´ "one hot"
random.shuffle(lst)
print(lst) # Ð½Ð°Ð¿ÐµÑ‡Ð°Ñ‚Ð°ÐµÐ¼ Ð¸Ð·Ð½Ð°Ñ‡Ð°Ð»ÑŒÐ½Ñ‹Ð¹ Ñ€Ð°Ð½Ð´Ð¾Ð¼Ð½Ð¾ ÑÐ³ÐµÐ½ÐµÑ€Ð¸Ñ€Ð¾Ð²Ð°Ð½Ð½Ñ‹Ð¹ ÑÐ¿Ð¸ÑÐ¾Ðº
for i in lst:
  if i == 'robot':
    i = 0
    lst_1.append(i)
  elif i == 'human':
    i = 1
    lst_1.append(i)
print(lst_1) # Ð½Ð°Ð¿ÐµÑ‡Ð°Ñ‚Ð°ÐµÐ¼ "one hot" ÑÐ¿Ð¸ÑÐ¾Ðº

data = pd.DataFrame({'whoAmI':lst_1}) # Ð²Ñ‹Ð²ÐµÐ´ÐµÐ¼ Ð¿Ð¾Ð»ÑƒÑ‡Ð¸Ð²ÑˆÑƒÑŽÑÑ Ñ‚Ð°Ð±Ð»Ð¸Ñ†Ñƒ
data.head(n=20)
```
